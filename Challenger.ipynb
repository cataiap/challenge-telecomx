# 6. Estandarización y transformación de datos (opcional)
La estandarización y transformación de datos es una etapa opcional, pero altamente recomendada, ya que busca hacer que la información sea más consistente, comprensible y adecuada para el análisis. # 9. Recuento de evasión por variables categóricas
Ahora, exploraremos cómo se distribuye la evasión según variables categóricas, como género, tipo de contrato, método de pago, entre otras.

Este análisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudará a orientar acciones estratégicas.

```python
# Variables categóricas a analizar en Churn
vars_cat = ["gender", "Contract", "PaymentMethod", "InternetService"]
```

```python
# Conteo de clientes por Contract y Churn (Yes/No)
conteo_contract = df_final.groupby(["Contract", "Churn"]).size().reset_index(name="Cantidad")
conteo_contract
```

```python
# Tabla dinámica: filas=Contract, columnas=Churn, valores=Cantidad
tabla_contract = pd.pivot_table(
    conteo_contract,
    index="Contract",
    columns="Churn",
    values="Cantidad",
    aggfunc="sum",
    sort=False
)

tabla_contract = tabla_contract.fillna(0)
tabla_contract
```

```python
# Tabla dinámica: filas=Contract, columnas=Churn, valores=Cantidad
tabla_contract = pd.pivot_table(
    conteo_contract,
    index="Contract",
    columns="Churn",
    values="Cantidad",
    aggfunc="sum",
    sort=False
)

tabla_contract = tabla_contract.fillna(0)
tabla_contract
```

```python
# Total por categoría y tasa de churn (%)
tabla_contract["Total"] = tabla_contract.sum(axis=1)
tabla_contract["Churn_rate_%"] = (tabla_contract["Yes"] / tabla_contract["Total"]) * 100

# Ordenamos por mayor churn
tabla_contract = tabla_contract.sort_values("Churn_rate_%", ascending=False)

tabla_contract
```

```python
tablas_cat = {}

for v in vars_cat:
    conteo = df_final.groupby([v, "Churn"]).size().reset_index(name="Cantidad")

    tabla = pd.pivot_table(
        conteo,
        index=v,
        columns="Churn",
        values="Cantidad",
        aggfunc="sum",
        sort=False
    ).fillna(0)

    tabla["Total"] = tabla.sum(axis=1)
    tabla["Churn_rate_%"] = (tabla["Yes"] / tabla["Total"]) * 100
    tabla = tabla.sort_values("Churn_rate_%", ascending=False)

    tablas_cat[v] = tabla

    print(f"\n--- Churn por {v} ---")
    display(tabla)
```

```python

df_plot = tablas_cat["Contract"].reset_index()

fig, ax = plt.subplots(figsize=(8, 2))
sns.barplot(data=df_plot, y="Contract", x="Churn_rate_%", palette="tab10", orient="h", ax=ax)

ax.set(
    title="Tasa de churn (%) por tipo de contrato",
    xlabel="Churn rate (%)",
    ylabel="Contract"
)
plt.show()
```

Por lo mismo, comprender **qué perfiles y condiciones están más asociadas al Churn** es clave para diseñar estrategias de retención efectivas.

En este notebook se desarrolló un proceso completo de **Extracción, Transformación, Limpieza y Análisis Exploratorio (EDA)**, con el objetivo de transformar datos en información accionable para el equipo de Data Science y para áreas de negocio (retención, marketing, operaciones y atención al cliente).

### 1.3 Metodología (resumen del flujo realizado)
El trabajo se organizó en etapas, manteniendo un enfoque paso a paso:

- **Extracción (E):** lectura del JSON desde la fuente y carga a pandas.

### 3.1 Distribución general de Churn

Primero se analizó la distribución de la variable objetivo **Churn** para entender el tamaño del problema.

Este gráfico permite ver rápidamente si la evasión es un caso minoritario o un problema significativo.

### 3.2 Churn por variables categóricas

Después se analizó cómo cambia el churn según variables categóricas (por ejemplo: género, tipo de contrato, método de pago, tipo de internet).

**¿Qué hicimos?**
1) Para cada categoría (por ejemplo, cada tipo de contrato), contamos:
   - cuántos clientes tienen `Churn = No`
   - cuántos tienen `Churn = Yes`
2) Calculamos la **tasa de churn (%)** por categoría:
   
   \[
   \text{Churn rate (%) } = \frac{\text{Yes}}{\text{Total}} \times 100
   \]

**¿Qué significa este resultado?**
- Una categoría con churn rate alto es un **grupo de riesgo**, donde más clientes tienden a cancelar.
- Esto permite priorizar acciones en segmentos específicos (por ejemplo, un contrato o método de pago con churn muy alto).

A continuación se muestran tablas (No/Yes/Total) y gráficos para comparar tasas por categoría.

```python

vars_cat = ["gender", "Contract", "PaymentMethod", "InternetService"]

tablas_cat = {}

for v in vars_cat:
    conteo = df_final.groupby([v, "Churn"]).size().reset_index(name="Cantidad")

    tabla = pd.pivot_table(
        conteo,
        index=v,
        columns="Churn",
        values="Cantidad",
        aggfunc="sum",
        sort=False
    ).fillna(0)

    tabla["Total"] = tabla.sum(axis=1)
    tabla["Churn_rate_%"] = (tabla["Yes"] / tabla["Total"]) * 100
    tabla = tabla.sort_values("Churn_rate_%", ascending=False)

    tablas_cat[v] = tabla

    print(f"\n--- Churn por {v} ---")
    display(tabla)
```

```python

df_plot = tablas_cat["Contract"].reset_index()

fig, ax = plt.subplots(figsize=(8, 3))
sns.barplot(data=df_plot, y="Contract", x="Churn_rate_%", palette="tab10", orient="h", ax=ax)

ax.set(
    title="Tasa de churn (%) por tipo de contrato",
    xlabel="Churn rate (%)",
    ylabel="Contract"
)
plt.show()
```
